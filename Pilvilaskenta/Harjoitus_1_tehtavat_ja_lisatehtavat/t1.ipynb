```python
from keras.models import load_model
from PIL import Image, ImageOps
import numpy as np
from google.colab import drive
drive.mount('/content/drive')
# Load the model
model = load_model('/content/drive/MyDrive/Pilvilaskenta2021_Noora_Angelva/keras_model.h5')

```

```python
# Create the array of the right shape to feed into the keras model
# The 'length' or number of images you can put into the array is
# determined by the first position in the shape tuple, in this case 1.
data = np.ndarray(shape=(1, 224, 224, 3), dtype=np.float32)
# Replace this with the path to your image
image = Image.open('/content/drive/MyDrive/Pilvilaskenta2021_Noora_Angelva/testikuva.jpg')
#resize the image to a 224x224 with the same strategy as in TM2:
#resizing the image to be at least 224x224 and then cropping from the center
size = (224, 224)
image = ImageOps.fit(image, size, Image.ANTIALIAS)

# turn the image into a numpy array
image_array = np.asarray(image)
# Normalize the image
normalized_image_array = (image_array.astype(np.float32) / 127.0) - 1
# Load the image into the array
data[0] = normalized_image_array

# run the inference
prediction = model.predict(data)

# kaikkien kategorioiden tunnistaminen
print(prediction)

# haetaan suurimman arvon indeksi tunnistamistuloksessa, sekä suurin arvo
highest_index = prediction.argmax(axis=-1)[0]
highest_value = prediction.max()

# ladataan labels.txt sekä tehdään siitä lista, erotus rivinvaihdolla
my_file = open("/content/drive/MyDrive/Pilvilaskenta2021_Noora_Angelva/labels.txt", "r")
content = my_file.read()
class_names = content.split("\n")

print("\n\n\n")

# tulostetaan tulokset
print("Todennäköisin kategoria:")
print(class_names[highest_index])
print()

# tulostetaan korkein todennäköisyys ja kerrotaan 100:lla, eli prosenteiksi
print("Todennäköisyys: ")
print(f"{round(highest_value * 100, 1)}%")  
```
